### 퍼셉트론 - 이진 분류

```python
import numpy as np
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

```python
class Perceptron:
    def __init__(self, num_feature, learning_rate):
#         self.w = np.random.uniform(size=num_feature+1)
        self.w = np.zeros((num_feature+1,))
        self.eta = learning_rate
        self.errors = []
        
    def forward(self, x):
        z = np.dot(x, self.w[1:]) + self.w[0]
        return self.step_function(z)
    
    @staticmethod
    def step_function(z):
        return np.where(z >= 0., 1, -1)
    
    def per_sample_training(self, sample, target):
        pred = self.forward(sample)
        error = target - pred
        delta = self.eta * error * np.concatenate([(1.,), sample], axis=0)
        self.w += delta
        return error
        
    def fit(self, X, y, iteration):
        num_sample = X.shape[0]
        print('init weights:', self.w)
        for epoch in range(iteration):
            print(f'{epoch} epoch start:')
            epoch_error = 0
            for sample, target in zip(X, y):
                error = self.per_sample_training(sample, target)
                epoch_error += int(error != 0)
                if error != 0:
                    print('new weights:', self.w)
            self.errors.append(epoch_error)
            print(f'{epoch} epoch - error: {epoch_error}/{num_sample}')
```

### 퍼셉트론 - 다중 클래스 분류

```python
X = []
y = []
colors = ['red', 'blue', 'green']
for i, (loc_0, loc_1) in enumerate([[1., 1.], [3., 1.], [2., 2]]):
    x0 = np.random.normal(loc=loc_0, scale=.2, size=50)
    x1 = np.random.normal(loc=loc_1, scale=.2, size=50)
    X.append(np.stack([x0, x1], axis=-1))
    y.append(np.ones(50) * i)
    plt.scatter(x0, x1, color=colors[i-1], label=str(i))
plt.legend(loc='upper left')
plt.show()
X = np.concatenate(X, axis=0)
y = np.concatenate(y)
print(X.shape, y.shape)
```

```python
class Perceptron:
    def __init__(self, num_feature, learning_rate):
        self.w = np.random.uniform(size=num_feature+1)
        self.eta = learning_rate
        self.errors = []
        
    def z(self, x):
        z = np.dot(x, self.w[1:]) + self.w[0]
        return z
        
    def forward(self, x):
        z = self.z(x)
        return self.step_function(z)
    
    @staticmethod
    def step_function(z):
        return np.where(z >= 0., 1, -1)
    
    def per_sample_training(self, sample, target):
        pred = self.forward(sample)
        error = target - pred
        delta = self.eta * error * np.concatenate([(1.,), sample], axis=0)
        self.w += delta
        return error
        
    def fit(self, X, y, iteration):
        num_sample = X.shape[0]
        for epoch in range(iteration):
#             print(f'{epoch} epoch start:')
            epoch_error = 0
            for sample, target in zip(X, y):
                error = self.per_sample_training(sample, target)
                epoch_error += int(error != 0)
            self.errors.append(epoch_error)
#             print(f'{epoch} epoch - error: {epoch_error}/{num_sample}')
```

```python
perceptron_dict = {}
for i in range(3):
    perceptron = Perceptron(2, .1)
    perceptron.fit(X, np.where(y==i, 1, -1), 50)
    perceptron_dict[i] = perceptron
    plt.plot(range(1, len(perceptron.errors)+1), perceptron.errors)
    plt.title(f'{i} perceptron errors')
    plt.xlabel('epochs')
    plt.ylabel('errors')
    plt.show()
```

```python
def plot_decision_regions(X, y, classifier, resolution=.02):
    markers = ('s', 'x', 'o')
    colors = ('red', 'blue', 'lightgreen')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.forward(np.array([xx1.flatten(), xx2.flatten()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        cl_sample = X[y==cl]
        plt.scatter(x=cl_sample[..., 0], y=cl_sample[..., 1], alpha=.8, c=colors[idx], marker=markers[idx], label=cl, edgecolor='black')
    
    plt.xlabel('x0')
    plt.ylabel('x1')
    plt.legend(loc='upper left')
    plt.show()
```

```python
for i in range(3):
    plt.figure(figsize=(4, 4))
    plt.title(str(i))
    plot_decision_regions(X, np.where(y==i, 1, -1), perceptron_dict[i])
```

```python
class MulticlassClassification:
    def __init__(self, perceptron_dict):
        self.cls_names, self.cls_perceptrons = [], []
        for cls_name, cls_perceptron in perceptron_dict.items():
            self.cls_names.append(cls_name)
            self.cls_perceptrons.append(cls_perceptron)
            
    def forward(self, X):
        pred = []
        for perceptron in self.cls_perceptrons:  # 모든 퍼셉트론 이용
            z = perceptron.z(X)
            pred.append(z)
        z_stack = np.stack(pred)
        argmax = np.argmax(z_stack, axis=0)  # 퍼셉트론 결과 중 가장 큰 값이 나타내는 클래스를 선택한다.
        return argmax
```

```python
multi_cls = MulticlassClassification(perceptron_dict)
```

```python
plt.figure(figsize=(4, 4))
plot_decision_regions(X, y, multi_cls)
```
